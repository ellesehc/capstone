```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('processed_data.csv')
```

```python
data.info()
```

```python
data.drop(data.columns[[0,1,2,3,4,10]], axis = 1, inplace = True)
data.info()
```

```python
data.drop(data.columns[[0,2,5,6,7,8]], axis = 1, inplace = True)
data.info()
```

```python
data.info()

ph_data = data[data['Country'] == 'PH']
```

```python
ph_data.drop(['Country'], axis = 1, inplace = True)
ph_data.info()
```

```python
ph_data['Review Count'].value_counts()
ph_data.head(10)
```

```python
ph_data.drop(10, axis = 0, inplace = True)
```

```python
ph_data.head(10)
```

```python
review_count = []

for row in ph_data['Review Count']:
    review_count.append(int(row))
    
ph_data['# Reviews'] = review_count
ph_data['Target Rating'] = ph_data['Rating']
```

```python
ph_data.info()
```

```python
ph_data.drop(['Review Count'], axis = 1, inplace = True)
ph_data.drop(['Rating'], axis = 1, inplace = True)
```

```python
ph_data.iloc[:, 0:-1]
```

```python
ph_data.info()
```

```python
X = ph_data.iloc[:, 0:-1].values
y = ph_data.iloc[:, -1].values

ph_data.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error, mean_absolute_error, precision_score, recall_score, f1_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy: {}%".format(accuracy_score(y_test, y_pred) *100 ))
print("Precision Score: {}".format(precision_score(y_test, y_pred, average = 'weighted')))
print("Recall Score: {}".format(recall_score(y_test, y_pred, average = 'weighted')))
print("F1 Score: {}".format(f1_score(y_test, y_pred, average = 'weighted') ))
print("Mean Squared Error: {}".format(mean_squared_error(y_test, y_pred) ))
print("Root Mean Squared Error: {}".format(mean_squared_error(y_test, y_pred, squared = False)))
print("Mean Absolute Error: {}".format(mean_absolute_error(y_test, y_pred) ))
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error, mean_absolute_error, precision_score, recall_score, f1_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy: {}%".format(accuracy_score(y_test, y_pred) *100 ))
print("Precision Score: {}".format(precision_score(y_test, y_pred, average = 'weighted')))
print("Recall Score: {}".format(recall_score(y_test, y_pred, average = 'weighted')))
print("F1 Score: {}".format(f1_score(y_test, y_pred, average = 'weighted') ))
print("Mean Squared Error: {}".format(mean_squared_error(y_test, y_pred) ))
print("Root Mean Squared Error: {}".format(mean_squared_error(y_test, y_pred, squared = False)))
print("Mean Absolute Error: {}".format(mean_absolute_error(y_test, y_pred) ))
```
