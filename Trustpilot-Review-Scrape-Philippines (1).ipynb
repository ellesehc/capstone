```python
def get_last_page(url):
    
    string = ""
    try:
        for item in url.find_all("a", attrs = {'data-pagination-button-last-link': 'true'}):
            return int(string + item.get_text())
    except:
        return 1

#get_last_page(soup)
```

```python
def get_business_name(url):
    
    for item in soup.find_all('span', class_ = 'typography_typography__QgicV typography_h1__Xmcta typography_weight-heavy__E1LTj typography_fontstyle-normal__kHyN3 styles_displayName__GElWn'):
        return item.get_text().replace('\xa0', "")
```

```python
def get_average_rating(url):
    for item in soup.find_all('p', class_ = 'typography_typography__QgicV typography_bodysmall__irytL typography_color-gray-7__9Ut3K typography_weight-regular__TWEnf typography_fontstyle-normal__kHyN3'):
        string = ""
        return string + item.get_text()
```

```python
def get_review_grade(url):
    for item in soup.find_all('span', class_ = 'typography_typography__QgicV typography_bodysmall__irytL typography_color-gray-7__9Ut3K typography_weight-regular__TWEnf typography_fontstyle-normal__kHyN3 styles_text__W4hWi'):
        return item.get_text().strip().replace('\xa0', "").split("â€¢")[1].strip()
```

```python
def get_title(url):
    result = []
    
    for item in url.find_all("a", attrs = {'data-review-title-link':'true'}):
  
        result.append(item.get_text())
    
    return result

#get_title(soup)
```

```python
def get_review(url):
    result = []
    
    for item in soup.findAll('article', attrs = {'data-service-review-card-paper': 'true'}):
            try:
                result.append(item.p.text)
                #print(item.p.text)
            except:
                result.append("[BLANK]")
        
    return result

#get_review(soup)
```

```python
def get_rating(url):
    result = []
    
    for item in soup.find_all("div", class_= "styles_reviewHeader__iU9Px"):
        string = ""
        rating = item.img['alt'][6:8]
        string = string + rating
        result.append(string)
        


    return result

#get_rating(soup)
```

```python

def get_time(url):
    result = []
    
    for item in soup.find_all('div', class_='styles_reviewHeader__iU9Px'):
        result.append(item.time.text)
    
    return result

#get_time(soup)
```

```python
def get_review_count(url):
    result = []
    
    for item in soup.findAll('span', attrs = {'data-consumer-reviews-count-typography': 'true'}):
        result.append(int(item.get_text()[0]))
        
    return result

#get_review_count(soup)
```

```python
def get_country(url):
    result = []
    
    for item in soup.findAll('span', attrs = {'data-consumer-country-typography': 'true'}):
        result.append(item.get_text())
        
    return result

#get_country(soup)
    
```

```python
def get_status(url):
    
    result = []
    
    for item in soup.find_all('div', class_='styles_reviewHeader__iU9Px'):
        try:
            result.append(item.span.text)
        except:
            result.append('[BLANK]')
    
    return result
            
#get_status(soup)
```

```python
def get_content(URL):
    r = requests.get(url, headers = HEADERS)
    return r.text

def html(URL):
    html_content = get_content(URL)
    soup = BeautifulSoup(html_content, 'html.parser')
    return soup

business_name = []
average_rating = []
review_grade = []

title = []
review = []
rating = []
time = []
count = []
country = []
status = []


for site in websites:
    url = site
    prepend = 'http://webcache.googleusercontent.com/search?q=cache:'
    page_count = 1
    
    urls = []
    

    
    if get_last_page(html(url)) is None:
        for i in urls:
            r = requests.get(i, headers = HEADERS)
            html_content = r.text
            soup = BeautifulSoup(html_content, 'html.parser')

            title.append(get_title(soup))
            review.append(get_review(soup))
            rating.append(get_rating(soup))
            time.append(get_time(soup))
            count.append(get_review_count(soup))
            country.append(get_country(soup))
            status.append(get_status(soup))
            bn = [get_business_name(soup)] * len(get_title(soup))
            ar = [get_average_rating(soup)] * len(get_title(soup))
            rg = [get_review_grade(soup)] * len(get_title(soup))
            business_name.append(bn)
            average_rating.append(ar)
            review_grade.append(rg)
            

    else:
        while page_count <= get_last_page(html(url)):
            urls.append(url + str(page_count))
            page_count += 1

        for i in urls:
            r = requests.get(i, headers = HEADERS)
            html_content = r.text
            soup = BeautifulSoup(html_content, 'html.parser')

            title.append(get_title(soup))
            review.append(get_review(soup))
            rating.append(get_rating(soup))
            time.append(get_time(soup))
            count.append(get_review_count(soup))
            country.append(get_country(soup))
            status.append(get_status(soup))
            bn = [get_business_name(soup)] * len(get_title(soup))
            ar = [get_average_rating(soup)] * len(get_title(soup))
            rg = [get_review_grade(soup)] * len(get_title(soup))
            business_name.append(bn)
            average_rating.append(ar)
            review_grade.append(rg)
            

    

print("done!")
```

```python
def list_elements(l):
    result = []
    
    for element in l:
        for unit in element:
            result.append(unit)
    
    return result


data = {
        'Business Name': list_elements(business_name),
        'Average Rating': list_elements(average_rating),
        'Review Grade': list_elements(review_grade),
        'Title': list_elements(title),
        'Review': list_elements(review),
        'Rating': list_elements(rating),
        'Time': list_elements(time),
        'Review Count': list_elements(count),
        'Country': list_elements(country),
        'Status': list_elements(status)
       }
    
df = pd.DataFrame(data)
df.to_csv('trustpilot_reviews.csv')
```
